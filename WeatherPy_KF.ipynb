```python
#import dependencies
import csv
import numpy as np
import pandas as pd
import requests
import seaborn as sns
import matplotlib.pyplot as plt
from config import api_key
from citipy import citipy
import random
import json
```

```python
lat = []
lon = []
for x in range(1000):
    lat.append(random.randint(-90,90))
    lon.append(random.randint(-180,180))

coordinates =[]
coordinates = tuple(zip(lat, lon))
print(coordinates)
```

```python
cities = []
for coordinate_pair in coordinates:
    lat, lon = coordinate_pair
    cities.append(citipy.nearest_city(lat, lon))
#print(cities)
locations = []
for city in cities:
    name = city.city_name
    country_code = city.country_code
    locations.append({"name" : name
                     ,"country code" : country_code})
    #name 
    #id
    #lat
    #lon
    #while length 
df = pd.DataFrame(locations)
df.head(100)
#df.nunique()
```

```python
api_url = 'http://api.openweathermap.org/data/2.5/weather?q='
#change to city
#lat={lat}&lon={lon}''
units = 'imperial'
#city = 'Dublin,IE'
urlend = '&appid={}&units={}'.format( api_key, units)
```

```python
city_name = []
city_country = []
city_lat = []
city_temp = []
city_humid = []
city_cloudiness = []
city_wind = []
i = 0
for i in range(5):
    response = requests.get(api_url + locations[i]+ urlend).json()
    #print(json.dumps(response, indent=1, sort_keys=False))
    city_name.append(response['name'])
    city_country.append(response['sys']['country'])
    city_lat.append(response['coord']['lat'])
    city_temp.append(response['main']['temp'])
    city_humid.append(response['main']['humidity'])
    city_cloudiness.append(response['clouds']['all'])
    city_wind.append(response['wind']['speed'])
```

```python
city_dict = {
    'City':city_name,
    'Country_Code': city_country,
    'Latitude':city_lat,
    'Tempurature':city_temp,
    'Humidity':city_humid,
    'Cloudiness_%':city_cloudiness,
    'Wind_(mph)':city_wind 
}
city_data = pd.DataFrame(city_dict)
city_data.head()
```
